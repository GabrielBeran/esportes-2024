```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
```

```python
websites = ['https://en.volleyballworld.com/volleyball/competitions/volleyball-nations-league/teams/men/', 'https://en.volleyballworld.com/volleyball/competitions/volleyball-nations-league/teams/women/']
links = []
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

for website in websites:
    # Abra a página da web
    driver.get(website)
    driver.maximize_window()

    # Encontre todos os elementos <a> dentro dos elementos <div>
    elementos_a = driver.find_elements(By.XPATH, '//a[@class="d3-l-col__col-2"]')

    # Itere sobre os elementos <a>
    for elemento_a in elementos_a:
        # Obtenha o valor do atributo "href" do elemento <a>
        link = elemento_a.get_attribute('href')

        # Adicione o link à lista de links
        links.append(link)
```

```python
links = [link.replace('schedule', 'players') for link in links]
print(links)
len(links)
```

```python
# Base URL
base_url = "https://en.volleyballworld.com"

jogadores = []

for link in links:
    driver.get(link)
    time.sleep(5)  # Aguarde o carregamento da página, ajuste conforme necessário
    
    # Usar BeautifulSoup para analisar o conteúdo da página
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Encontrar todos os tr com a classe especificada (e todas as variações)
    row_classes = [f'vbw-o-table__row vbw-o-table__row--{i}' for i in range(100)]  # Ajuste o range conforme necessário
    
    for row_class in row_classes:
        trs = soup.find_all('tr', class_=row_class)
        for tr in trs:
            td = tr.find('td', class_='vbw-o-table__cell playername')
            if td:
                a_tag = td.find('a', class_='d3-l-col__col-2', href=True)
                if a_tag:
                    full_url = base_url + a_tag['href']
                    jogadores.append(full_url)

# Exibir os links coletados
for jogador in jogadores:
    print(jogador)
```

```python
# Lista para armazenar os nomes dos jogadores
nome_jogadores = []

for jogador in jogadores:
    driver.get(jogador)
    time.sleep(2)
    
    # Usar BeautifulSoup para analisar o conteúdo da página
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Encontrar o h3 com a classe especificada
    h3_tag = soup.find('h3', class_='vbw-player-name')
    if h3_tag:
        nome_jogadores.append(h3_tag.text.strip())

# Exibir os nomes coletados
for nome in nome_jogadores:
    print(nome)
```

```python
df_list = pd.DataFrame({'Membro': nome_jogadores})
df_list.to_csv('Membros Nation League de Vôlei_2024.csv', index=False)
```
