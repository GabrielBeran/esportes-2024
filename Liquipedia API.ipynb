```python
import requests
import json

# Função para buscar o link da página de um time específico no contexto de Dota 2
def buscar_pagina_time(time_nome):
    url_busca = f"https://liquipedia.net/dota2/api.php?action=query&list=search&srsearch={time_nome}&format=json"
    response = requests.get(url_busca)
    data = response.json()

    # Exibir o JSON completo da resposta
    print("Resposta JSON completa:")
    print(json.dumps(data, indent=4))

    # Verifique se a lista de resultados contém itens
    if 'query' in data and 'search' in data['query'] and len(data['query']['search']) > 0:
        pagina_titulo = data['query']['search'][0]['title']
        return pagina_titulo
    else:
        print("Nenhuma página encontrada para o time fornecido.")
        return None

# Nome do time
time_nome = "Team Spirit"

# Buscar a página do time no contexto de Dota 2
pagina_titulo = buscar_pagina_time(time_nome)

# Exibir o título da página encontrada
if pagina_titulo:
    print(f"Página encontrada: {pagina_titulo}")
```
